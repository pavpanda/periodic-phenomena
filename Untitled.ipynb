```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ["Heart beat", "Solar eclipse", "SAIL Amsterdam", "Halley's Comet", "Transit of Venus", "Magnetic Field Reversal"]
y = [0.0000000317057705, 1.5, 5, 75.5, 110, 250000]
plt.style.use('ggplot')
plt.xticks(rotation=60)
plt.bar(x, np.log(y))
plt.title("Log of time between occurrences vs Event type")
plt.ylabel("Log of time between occurences")
plt.xlabel("Event type")
plt.savefig("all.png", bbox_inches="tight")
```
